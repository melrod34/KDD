```python
import pandas as pd 
import numpy as np 
import plotly
import mlxtend
```

```python
titanic = pd.read_csv("Titanic.csv")
titanic.shape
```

```python
# Gather All Items of Each Transactions into Numpy Array
transaction = []
for i in range(0, titanic.shape[0]):
    for j in range(0, titanic.shape[1]):
        transaction.append(titanic.values[i,j])
# converting to numpy array
transaction = np.array(transaction)
print(transaction)
```

```python
#  Transform Them a Pandas DataFrame
df = pd.DataFrame(transaction, columns=["items"])
# Put 1 to Each Item For Making Countable Table, to be able to perform Group By
df["incident_count"] = 1
df
```

```python
# Making a New Appropriate Pandas DataFrame for Visualizations
df_table = df.groupby("items").sum().sort_values("incident_count", ascending=False).reset_index()
df_table
```

```python
# Transform Every Transaction to Seperate List & Gather Them into Numpy Array
transaction2 = []
for i in range(titanic.shape[0]):
    transaction2.append([str(titanic.values[i,j]) for j in range(titanic.shape[1])])
```

```python
transaction2 = np.array(transaction2)
transaction2
```

```python
# importing the required module
from mlxtend.preprocessing import TransactionEncoder
# initializing the transactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transaction2).transform(transaction2)
te_ary
```

```python
dataset2 = pd.DataFrame(te_ary, columns=te.columns_)
# dataset after encoded
dataset2
```

```python
# select top 30 items
first30 = df_table["items"].head(30).values
# Extract Top 30
dataset3 = dataset2.loc[:,first30]
# shape of the dataset
dataset3.shape
```

```python
#Importing Libraries
from mlxtend.frequent_patterns import fpgrowth
#running the fpgrowth algorithm
res=fpgrowth(dataset3,min_support=0.1, use_colnames=True)
res.head(10)
```

```python
# importing required module
from mlxtend.frequent_patterns import association_rules
# creating asssociation rules
rules=association_rules(res, metric="confidence", min_threshold=0.7)
# printing association rules
rules 
```

```python

final_rules = rules[rules['consequents'].apply
                       (lambda x: len(x) == 1 and 
                        ('Yes' in x or 'No' in x))]
final_rules
```
