```python
import pandas as pd
from sklearn.decomposition import PCA
import scipy
from scipy import stats
import numpy as np
```

```python
data = pd.read_table("Nichols_Nicki.dat")
data
```

```python
df = pd.DataFrame(data)
variance = df.var()
sorted_variance = variance.sort_values()
print(sorted_variance.head())
df2 = data.drop(columns=['ias34','ias23'])
```

```python
matrix = df2.corr()
print(matrix)
```

```python
sums = matrix.sum()
min_corr = sums.nsmallest(3).index
print(min_corr)
```

```python
new_df = df2.drop(columns=['ias13','ias22', 'ias32'])
print("5 columns were dropped, 2 due to low variances and 3 due to low correlations.")
print("Columns ias23 and ias34 were dropped for low variances.")
print("Columns ias13, ias22 and ias32 were dropped due to low correlations.")
```

```python
new_df
```

```python
cov_mat = np.cov(new_df.T)
eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)
eigen_vals
```

```python
pca = PCA(n_components=None)
data_pca = pca.fit_transform(new_df)
pca.explained_variance_ratio_
```
