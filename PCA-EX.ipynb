```python
import pandas as pd
```

```python
raq = pd.read_table('raq.dat')
```

```python
raq
```

```python
from sklearn.decomposition import PCA
```

```python
#Barlett's test
import scipy
from scipy import stats
scipy.stats.bartlett(*raq.values)
```

```python
import numpy as np
```

```python
cov_mat = np.cov(raq.T)
```

```python
cov_mat
```

```python
#Calculating the eigen value
eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)
```

```python
eigen_vals
```

```python
pca = PCA(n_components=None)
```

```python
raq_pca = pca.fit_transform(raq)
```

```python
pca.explained_variance_ratio_
```

```python
pca = PCA(n_components=5)
```

```python
raq_pca2 = pca.fit_transform(raq)
```

```python
pca.explained_variance_ratio_
```

```python
raq_pca2
```

```python
pca = PCA(n_components=2)
```

```python
raq_pca3 = pca.fit_transform(raq)
```

```python
raq_pca3
```

```python
raq_pca3_dataframe = pd.DataFrame(data = raq_pca3
             , columns = ['principal component 1', 'principal component 2'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16,10))
sns.scatterplot(
    x="principal component 1", y="principal component 2",
    data=raq_pca3_dataframe,
    legend="full",
    alpha=0.3
)
```
